# Seaborn
é uma biblioteca para gráficos estátisticas em python. Integrado com matplotlib e pandas.

Ele ajuda a explorar e entender os dados. Podemos usar arrays e dataframes.

## Introdução

## exemplo:

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

## Visualização

### relplot()
Serve para visualizar diferentes relações estástitisca. Enquanto o gráfico de dispersão geralmente são efetivos, relações ode uma variavel representa tempo, é melhor representado por linha. Podemos passar o parâmetro **kind** para mudar as representações.

Tanto para os gráficos de dispersão, quanto para linha, podemos passar os parâmetros **size** e **style**, mas afeta de forma diferente:



```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```

### Estimativa
O seaborn pode fazer estimativas estátisticas:

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

### lmplot()
Podemos usar o modelo da regressão linear:

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

### Histográmas
Podemos analisar a distribuição dos ados usando um histograma, que pode ser feito usando **displot()**.

parâmetro:

o primeiro são os dados

hist - booleano - mostrar as barras

kde - booleano - mostrar a linha de densidade

bins - número de barras

color - você sabe

hist_kws - json

    edgecolor: cor da borda
    
col = como você quer dividir

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

Searboarn possui várias formas:

```python
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker", rug=True)
```

## Plots especializado para dados categorizados

#### catplot()

Esses plots possui diferentes níveis de granularidade

```python
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```

Podemos usar a distribuição usando o 'violin'

```python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker", split=True)
```

Podemos ver apenas as média

```python
sns.catplot(data=tips, kind="bar", x="day", y="total_bill", hue="smoker")
```

### Views compostas para dados multivalorados

Podemos juntar vários tipos de plots para criar um resumo rápidos dos dados. Um deles é o **jointplot()**, concentra-se em um único relacionamento. Representando a distribuição conjunta entre duas variáveis junto com a distribuição marginal:

```python
penguins = sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

Podemos usar o **pairplot()** tem uma visão mais ampla: mostra as distruições conjuntas e margianis de todas as relações de pares e para cada variável, respectivamente

```python
sns.pairplot(data=penguins, hue="species")
```

```python
sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g"
)
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", kind="hist")
```
