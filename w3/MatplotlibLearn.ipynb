## Plotando pontos
#### plot(x-axis, y-axis)
desenha pontos no gráfico

Por padrão, plot() desenha uma linha que passa pelos pontos

x-axis: contém as abscissas (x) dos pontos
y-axis: contém as ordenadas (y) dos pontos

```python
xpoints = np.array([0, 1 , 2])
ypoints = np.array([0,10,-10])

plt.plot(xpoints, ypoints)
plt.show()
```

Imprimindo um poligono de n vértices

### plotando sem a linha
Para plotar sema  linha, passa como o parâmetro notation o valor 'o'

```python
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

### Valores padrão para x
Se não for especificado um valor para x, é passado como padrão os valores 0, 1, 2, 3...

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

### Marcadores
Para enfatiza os pontos com um circulo, passe 'o' como parâmetro marker

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

Para enfatizar com uma estrela

```python
plt.plot(ypoints, marker = '*')
plt.show()
```

Existem outras opções, que podem ser vistas a seguir:

```python
markers = ('o', '*', '.', ',', 'x', 'X', '+', 'P', 's', 'D', 'd', 'p', 'H', 'h', 'v', '^', '<', '>', '1', '2', '3', '|', '_')

x = 0
y = 0
for m in markers:
    plt.plot([x], [y], marker = m)
    y += 1
    
    if(y == 4):
        x += 1
        y = 0
```

## formatar
Para formatar, devemos usar os parâmetros fmt
"marker|line|color"

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

Na parte do 'line', existe 4 opções:
    '-' Linha sólida
    ':' Linha pontilhada
    '--' linha tracejadas
    '-.' linha pontilha e tracajada

```python
lineSytax = ('-', ':', '--', '-.')

for i, l in enumerate(lineSytax):
    ypoints = np.array([0, 10])
    plt.plot([i, i],ypoints, f'o{l}r')
plt.show()
```

Cores:
'r': Vermelho
'g': Verde
'b': Azul
'c': Ciano
'm': Margenta
'y' Amarelo
'k': Preto
'w' Branco

```python
cores = ('r', 'g', 'b', 'c', 'm', 'y', 'k', 'w')
for i, c in enumerate(cores):
    ypoints = np.array([0, 10])
    plt.plot([i, i],ypoints, f'o-{c}')
plt.show()
```

### Tamanho do marcador
Para especificar o tamanho do marcador, passe ele no argumento markersize, ou sua versão curta, ms

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

### Cor do marcador
Existem 2 formas: 
markeredgecolor ou mec = muda a cor da borda
markerfacecolor ou mfc = muda a cor do marcador

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc='g')
plt.show()
```

Você pode usar hexadecimal:

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

### Multipla linhas
Você pode plotar mais de uma vez

```python
n = 3
xaxis = []
yaxis = []

anguloDiff = 2 * pi / n
angulo = pi / 2

ax = plt.axes()
ax.set_aspect('equal')

for i in range(n):
    xaxis.append(cos(angulo))
    yaxis.append(sin(angulo))
    angulo += anguloDiff

xaxis.append(xaxis[0])
yaxis.append(yaxis[0])
    
plt.plot(xaxis, yaxis)
plt.show()
```

### Criar label e título

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

### Definir fonte
podemos usar o parâmetro fontdict no xlabel, ylabel e titúlo

No fontdict você pode colocar as chaves family, color e size

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

Você pode especificar o locla do título, com os valores 'left', 'right' ou 'center'

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

### Adcionar linhas de grade em um plot
Podemos usar a função grid() para adcionar as linhas no plot

```python
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

Você pode usar o parâmetro para especificar onde colocar a grade, 'x', 'y' ou 'both'

```python
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

Podemos personalizar as grades, com os parâmetros color, linestyle e linewidth

```python
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
```

## Gráfico de dispersão
Podemos usar a função scatter() para desenhar o gráfico de dispersão,
obtendo 2 parâmetros, um para x outro para y

```python
x = np.array([randint(0, 100) for _ in range(1000)])
y = np.array([randint(0, 100) for _ in range(1000)])

plt.scatter(x, y)
plt.show()
```

Comparar 2 plotsm por exemplo, comparar a idade e a velocidade de carros, em 2 dias.

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

### cores
O exemplo anterior teve 2 cores diferentes, porém é possível fazer isso manualmente, apartir do argumento color ou c

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')

plt.show()
```

Podemos especificar a cor de cada ponto, com um array

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)
```

## Colormap
Colormap é uma escala de cores, de 0 até 100

```python
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()
```

### Tamanho
Podemos mudar o tamanho com o argumento s

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)
```

Podemos especificar a transparência com o argumento alpha

```python
plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

## combinar cor, tamanho e alpha

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

## Gráfico de pizza
Podemos usar a função pie()
O primeiro parâmetro são os dados
O parâmetro labels é um array dos rótulos
Para especificar o ângulo de inicio, o parâmetro startangle, anti-horário

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.subplot(1,2,1)
plt.pie(y, labels=mylabels)
plt.subplot(1,2,2)
plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

Podemos destacar um ou mais fatias da pizza

```python
myexplode = [0.1, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

Podemos colocar sombras com shadow

```python
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

Podemos especificar as cores

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
Apesar de ter os labels, também é possivel colocar legenda com legend()
O parâmetro title serve para colocar um título
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title="Four Fruits:")
plt.show() 
```

## Boxplot

Para isso, podemo usar o plt.boxplot()

o primeiro parâmetro são os dados

Booleanos:

    vert = colocar na vertical
    
    showfliers = mostrar os dados discrepantes
    
    notch = entorta o quadrado, para representar um pouco a distribuição
    
    patch_artist = para pintar

```python
base = pd.read_csv('trees.csv')

plt.boxplot(base.Volume, vert=True, showfliers=True, notch=True, patch_artist=False)
plt.title("Árvores")
plt.xlabel("Volume")
```
