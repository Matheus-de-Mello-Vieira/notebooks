```python
import numpy as np
import matplotlib.pyplot as plt
import math as e
from scipy.stats import norm, probplot
```

```python
n = len(valores)
imagem = []
for i in range(1,n+1):
    imagem.append(round(norm.ppf((i - 0.5)/n), 2))
```

```python
plt.plot(imagem,valores, 'o')
```

```python
fig, ax = plt.subplots()
probplot(valores, fit=True, plot=ax)
plt.show()
```
