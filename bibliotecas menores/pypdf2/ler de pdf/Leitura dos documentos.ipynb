```python
import pathlib
import os
import pandas as pd
```

```python
from converter import read_file_set;
```

```python
import PyPDF2
import re

def read_file_set(path, mark):
    data = []
    for fn in (path).iterdir():
        with open(fn, 'rb') as file:
            read_pdf = PyPDF2.PdfFileReader(file)
            text = ""
            for i in range(read_pdf.getNumPages()):
                page = read_pdf.getPage(i)
                page_content = page.extractText()
                text += ' '.join(page_content)
            text = text.replace("Tribunal de Justiça de Minas Gerais","")
            text = re.sub(".*Data da Publicação:(EMENTA: *)?", "", text)
            data.append((text, mark))
    return data
```

```python
base_dir = pathlib.Path(os.getcwd())
```

```python
dados = read_file_set(base_dir/"Acordo",0)
dados += read_file_set(base_dir/"litigio",1)
```

```python
df = pd.DataFrame(data, columns=["text", "result"])
df = df[3:-3]
```

```python
df.to_csv("./dataframe.csv", index=False)
```
