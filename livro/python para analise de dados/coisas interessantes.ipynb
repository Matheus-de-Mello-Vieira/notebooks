```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels as sm
```

# Avaliar uma função $R^2 \to R$

```python
points = np.arange(-5,5,.01) # 1000 pontos igualmente espaçados
xs,ys = np.meshgrid(points, points) # produto cartesiano
z = np.sqrt(xs ** 2 + ys ** 2)
plt.imshow(z, cmap="gray_r", origin="lower", extent=(-5,5,-5,5))
plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
plt.show()
```

# Passeios aleatórios

```python
import random
position = 0
walk = [position]
steps = 100
for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)
plt.plot(walk)
```

```python
nsteps = 100
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
plt.plot(walk)
```

```python
# instante da primeira cruzada - o passo que o passeio aleatório alcança um determinado valor
(np.abs(walk) >= 10).argmax() # argmax retorna o primeiro valor máximo no array (True é 1)
```
