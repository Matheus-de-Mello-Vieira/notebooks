```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ex3 = np.array([[47, 62, 65, 70, 70, 78, 95, 100, 114, 118, 124, 127, 140, 140, 140, 150, 152, 164, 198, 221],
[38, 62, 53, 67, 84, 79, 93, 106, 117, 116, 127, 114, 134, 139, 142, 170, 149, 154, 200, 215]])
```

```python
plt.plot(*ex3,'o')
plt.show()
```

```python
X, Y = ex3
```

```python
X = X - X.mean()
Y = Y - Y.mean()
```

```python
plt.plot(X,Y,'o')
```

```python
theta = 1/2 * np.arctan(-2 * (X*Y).sum() / (X**2 - Y**2).sum())
```

```python
theta
```

```python
plt.plot(X,Y,'o')
plt.plot([0,np.sin(theta)*20],[0,np.cos(theta)*20])
```

```python
Xr = X * np.cos(-theta) - Y * np.sin(-theta)
Yr = X * np.sin(-theta) + Y * np.cos(-theta)
```

```python
plt.plot(Xr,Yr,'o')
plt.plot(X,Y,'o')
```
